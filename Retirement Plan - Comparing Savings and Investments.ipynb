# Retirement Plan

# Comparing monthly contributions

```python
from matplotlib import pyplot as plt

def retire(monthly, rate, terms):
    '''
    Takes 3 arguments and computes the compound interest
    
    monthly: monthly contribution
    rate: yearly interest rate
    terms: number of months before retirement
    '''
    savings = [0]
    base = [0]
    m_rate = rate/12 #annual rate divide by 12
    for i in range(terms):
        base += [i]
        #adding the monthly contribution to previous month's savings
        savings += [savings[-1] * (1 + m_rate) + monthly] 
    return base, savings  #x, y axis
    
    
#displaying results vs. month
def display_retire_monthlies(monthlies, rate, terms):
    plt.figure('retire_month')
    plt.clf()
    for monthly in monthlies:
        xvals, yvals = retire(monthly, rate, terms)
        plt.plot(xvals, yvals, label = f'retire: ${monthly}' )
        plt.legend(loc = 'upper left')
    plt.title(f'Total Savings after {terms//12} years')
    plt.xlabel('months'.title())
    
#checking the total savings for different monthly contributions ($500 - 1100) assuming monthly rate = 5% and term of 40 years
display_retire_monthlies([500, 600, 700, 800, 900, 1000, 1100], 0.05, 40*12)
```

# Comparing rate of growth on investments

```python
def display_retire_rates(month, rates, terms):
    plt.figure('retire_rate')
    plt.clf()
    for rate in rates:
        xvals, yvals = retire(month, rate, terms)
        plt.plot(xvals, yvals, label = 'retire: $' + str(month) + '/month @ ' + str(int(rate*100)) + str('%'))
        plt.legend(loc = 'upper left')
    plt.title(f'Total Savings after {terms//12} years')
    plt.xlabel('months'.title())

#checking the savings at different interest rates
display_retire_rates(800, [0.01, 0.03,0.05,0.07], 40*12)  
```
